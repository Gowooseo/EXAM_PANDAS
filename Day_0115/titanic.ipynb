```python
tiDF["당시생존여부"] = tiDF["당시생존여부"].astype("bool")
tiDF["사회적 지위"] = tiDF["사회적 지위"].astype("category")
tiDF["성별"] = tiDF["성별"].astype("category")
# tiDF["나이"].astype("int")
tiDF["형제/배우자 수"] = tiDF["형제/배우자 수"].astype("int8")
tiDF["부모/자녀 수"] = tiDF["부모/자녀 수"].astype("int8")
tiDF["클래스"] = tiDF["클래스"].astype("category")
tiDF["남/여/어린이"] = tiDF["남/여/어린이"].astype("category")
tiDF["승선도시"] = tiDF["승선도시"].astype("category")
tiDF.info()
```

```python
# 나이에 이상치가 있다!!
tiDF["나이"].unique() # 소수인 값들은 반올림을 하고 nan을 없애주자
```

```python
tiDF["나이"].mean() # 결측치에 넣어줄 평균값 
```

```python
tiDF["나이"].fillna(tiDF["나이"].mean(), inplace=True)
```
