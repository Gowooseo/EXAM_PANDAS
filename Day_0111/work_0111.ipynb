(9)DataFrame의 원소들 모두 출력, 단 2가지 방법 사용

### 방법1.iloc 사용 int 사용


```python
for row in range(movieDF.shape[0]): # shape는 (행의 개수,열의 개수)가 나오므로 shape[0]은 행의 개수.shape[1]은 열의개수가 출력된다.
    for col in range(movieDF.shape[1]):
        print(row,col,movieDF.iloc[row,col])  #movieDF.iloc[row,col]:row행 col 열에 있는 녀석들을 반복문에 따라 출력할 것
```

### 방법 2. loc 사용(레이블 사용)

```python
for row in movieDF.index:
    for col in movieDF.columns:
        print("----",col,movieDF.loc[row,col])
```
